```python
VLINES = np.arange(2012, 2023, 1)
import random
get_color = lambda n: ["#%06x" % random.randint(0, 0xFFFFFF) for _ in range(n)]
COLOR_SCALE = get_color(50)
print(COLOR_SCALE)
```

```python
fig, ax = plt.subplots(figsize = (14, 8.5))

# Background color
fig.patch.set_facecolor('#fafafa')
ax.set_facecolor('#fafafa')

# Vertical lines used as scale reference
for h in VLINES:
    ax.axvline(h, color='#e8e8e8', lw=0.6, zorder=0)

# Horizontal lines
ax.hlines(y=np.arange(0, 35), xmin=2012, xmax=2024, color='#e8e8e8', lw=0.6)

# Darker horizontal line at y=0
ax.hlines(y=0, xmin=2012, xmax=2024, color='#999999', lw=0.8)

for idx, group, in enumerate(df_mac["domain"].unique()):
    data = df_mac[df_mac["domain"] == group]
    color = COLOR_SCALE[idx]
    ax.plot("YEAR", "avg", color=color, data=data, label=group)
ax.legend(loc='upper right', bbox_to_anchor=(1.25, 1.0))
```
